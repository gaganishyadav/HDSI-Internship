```python
df1 = df.drop(columns = 'date')
df1
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

normalised_df1 = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)

feature_variable1 = normalised_df1['T2']
target_variable1 = normalised_df1['T6']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(feature_variable1, target_variable1, test_size=0.10, random_state=1)

x_train = x_train.values.reshape(-1, 1)
y_train = y_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)
y_test = y_test.values.reshape(-1, 1)

from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

linear_model.fit(x_train, y_train)

predicted_values = linear_model.predict(x_test)
```

```python
feature_variable1
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
df2 = df.drop(columns = ['date', 'lights'])
df2
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

normalised_df2 = pd.DataFrame(scaler.fit_transform(df2), columns=df2.columns)

feature_variable2_df = normalised_df2.drop(columns='Appliances')
target_variable = normalised_df2['Appliances']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(feature_variable2_df, target_variable, test_size=0.30, random_state=42)

from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()

linear_model.fit(x_train, y_train)

predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
#this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
```
